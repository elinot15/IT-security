# Esercizio 2

A stream cipher is an encryption algorithm that encrypts 1 bit or byte of plaintext at a time.
It uses an infinite stream of pseudorandom bits as the key. For a stream
cipher implementation to remain secure, its pseudorandom generator should be
unpredictable and the key should never be reused. The following file crypt.txt (inside
02_stream_chiper folder) has been encrypted using the most famous stream cypher
encoding algorithm with the key contained in file key.txt. Recover the original plaintext.

```python
encrypted=open("/home/alessia/Dropbox/Università/Sicurezza e privatezza/Lab1/SecLab1/02_stream_chiper/crypt.txt", "r")
encrypted=encrypted.read()

key=open("/home/alessia/Dropbox/Università/Sicurezza e privatezza/Lab1/SecLab1/02_stream_chiper/key.txt", "r")
key=key.read()
```

```python
key=key.encode()
encrypted=encrypted.encode()
decrypted=bytearray()
```

```python
for (k, val) in zip(key, encrypted): 
    decrypted.append(k^val)
```

```python
decrypted.decode()
```

# Esercizio 3


It is well-known that monoalphabetic substitution cipher (also known as monoalphabetic
cipher) is not secure, because it can be subjected to frequency analysis. In this lab, you are
given a cipher-text that is encrypted using a monoalphabetic cipher; namely, each letter in
the original text is replaced by another letter, where the replacement does not vary (i.e., a
letter is always replaced by the same letter during the encryption). Your job is to find out
the original text using frequency analysis and the encryption key. It is known that the
original text is an Italian poetry. We do keep the spaces between words, so you can still see
the boundaries of the words in the ciphertext. In real encryption using monoalphabetic
cipher, spaces will be removed. We also keep capital letters.
Guidelines: using the frequency analysis, you can find out the plaintext for some of the
characters quite easily. For those characters, you may want to change them back to itsplaintext, as you may be able to get more clues. It is better to use capital letters for plaintext,
so for the same letter, we know which is plaintext and which is ciphertext.

```python
ciphertext=open("/home/alessia/Dropbox/Università/Sicurezza e privatezza/Lab1/SecLab1/03_monoalphabetic_chiper/ciphertext.txt", "r")
ciphertext=ciphertext.read()
ciphertext=ciphertext.replace("\n", " ").strip()
ciphertext
```

```python
res = {i : ciphertext.count(i) for i in set(ciphertext) if(i.isalpha())} 
```

```python
freq=pd.DataFrame.from_dict(res, orient='index', columns=['Freq.'])
freq=freq.sort_values('Freq.', ascending=False)
freq=freq/len(ciphertext)*100
freq=freq[1:len(freq)]
#freq
```

```python
d_ita={'a':11.74, 'b':0.92, 'c':4.50, 'd':3.73, 'e':11.79, 'f':0.95, 'g':1.64, 'h':1.54, 'i':11.28, 'l':6.51, 'm':2.51, 'n':6.88, 'o':9.83, 'p':3.05, 'q':0.51, 'r':6.37, 's':4.98, 't':5.62, 'u':3.01,'v':2.10, 'z':0.49}
freq_ita=pd.DataFrame.from_dict(d_ita, orient='index', columns=['Freq.'])
freq_ita=freq_ita.sort_values('Freq.', ascending=False)
#freq_ita
```

```python
for i in range(0,6):
    ciphertext=ciphertext.replace(freq.index[i], (freq_ita.index[i]).upper()) 
ciphertext    
```

# Esercizio 4

All files recived from an external source can be corrupted, for example when they are
downloaded from the internet (due to transmission error), when a storage device faults or -in
the worst scenario- when an attackers change it's contents.
One way to ensure that programs or files have not been damaged or tampered is to
calculate their hash value and then comparing it with the checksum provided by the
source. If the hash matches, then you have a reasonable degree of confidence that the file
you downloaded is exactly the same as the one declared by it's source. If it is not, then the
file has been changed in some way.
Inside 04_integrity_checks folder there is a PDF and it’s pre-calculated hash, can you
check it’s integrity?

```python
md5=open("/home/alessia/Dropbox/Università/Sicurezza e privatezza/Lab1/SecLab1/04_integrity_checks/checksum.md5", "r")
md5=md5.read().strip()

sha1=open("/home/alessia/Dropbox/Università/Sicurezza e privatezza/Lab1/SecLab1/04_integrity_checks/checksum.sha1", "r")
sha1=sha1.read().strip()

sha256=open("/home/alessia/Dropbox/Università/Sicurezza e privatezza/Lab1/SecLab1/04_integrity_checks/checksum.sha256", "r")
sha256=sha256.read().strip()

sha512=open("/home/alessia/Dropbox/Università/Sicurezza e privatezza/Lab1/SecLab1/04_integrity_checks/checksum.sha512", "r")
sha512=sha512.read().strip()

pdf=open("/home/alessia/Dropbox/Università/Sicurezza e privatezza/Lab1/SecLab1/04_integrity_checks/PE_Format_Walkthrough.pdf", "rb")
pdf=pdf.read()

pdf
```

```python
hs.md5(pdf).hexdigest()==md5
```

```python
hs.sha1(pdf).hexdigest()==sha1
```

```python
hs.sha256(pdf).hexdigest()==sha256
```

```python
hs.sha512(pdf).hexdigest()==sha512
```

# Esercizio 5

Write a python script which is able to determine a value i such that H(i) =
00000xxxxxxxxxxxxxxxx, where H is the hash function SHA-1. The minimum number of
leading zeroes accepted is 3. The student has to provide the value of i.

```python
ALPHABET="qwertyuiopasdfghjklzxcvbnmQWERTYUIOPASDFGHJKLZXCVBNM0123456789"
LEN=4

def sha1_hash(num_seq):
        string=''.join(num_seq)
        hash_object=hs.sha1(string.encode())
        return hash_object.hexdigest()
```

```python
for word in it.product(ALPHABET, repeat=LEN):
        generated_hash=sha1_hash(word)
        if (generated_hash[0]=='0' and generated_hash[1]=='0' and generated_hash[2]=='0' and generated_hash[3]=='0' and generated_hash[4]=='0'):
            print('i:', word, 'gen:', generated_hash)
```

# Esercizio 6

Bruteforce attack is the most powerful cracking mode, it can try all possible character
combinations as passwords. However, cracking with exhaustive methods will never
terminate because of the number of combinations being too large. In this exercise we have
applied some constraints to get the job done in a reasonable time.
The folder 06_short_password_cracking contains some hashed passwords. Apply a brute
force attack in order to recover the passwords plaintexts, knowing that they are 5
characters long and contains only lowercase characters. The list of recovered passwords
has to be sent

```python
ALPHABET="abcdefghijklmnopqrstuvwxyz"
LEN=5
#data= open("/home/alessia/Dropbox/Università/Sicurezza e privatezza/Lab1/SecLab1/06_short_password_cracking/hacked_passwords.txt", "r")
#data=data.read()
data = pd.read_csv('/home/alessia/Dropbox/Università/Sicurezza e privatezza/Lab1/SecLab1/06_short_password_cracking/hacked_passwords.txt', sep=":", header = None)
data
```

```python
def md5_hash(char_sequence):
        string=''.join(char_sequence)
        hash_object=hs.md5(string.encode())
        return hash_object.hexdigest()
```

```python
for word in it.product(ALPHABET, repeat=LEN):
        generated_hash=md5_hash(word)
        if generated_hash in list(data[1]):
            print('The password of user', data[data[1]==generated_hash][0].to_string(index=False), 'is:', ''.join(word))
```
